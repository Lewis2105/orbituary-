Question 1: Designing a Way to Investigate Insider Threats
Analysis Methodology

If I worked on a possible insider threat using various types of data, I would use a step-by-step approach.

1:Data Integration and Normalization:

I would start by coming up with a way to combine network logs, access records and email metadata into one model. At this stage, timestamps, user identification and action descriptions are normalized for use in multiple systems.

For reviewing network logs, I would pay attention to attempts to log in, transfers of large amounts of data and rfssedes to confidential systems.

I would review all access records, including escalations, staff using access outside of work hours and workers accessing computer systems unrelated to what they do at work.

When dealing with emails, I would study how much I send, when and who to, instead of reading the contents to begin with, due to respecting privacy.

2:Behavioral Baseline Establishment:

Create a picture of how departments and individuals usually behave over a 3-6 month period by documenting behavior and activity.

Implement statistical procedures to know the typical ranges for login times, data amount accessed and ways of communicating.

3:Anomaly Detection:

In order to recognize deviations from the common trends, apply machine learning models like Isolation Forest and One-Class SVM.

Graph analysis can be used to find unusual activity in network communication or data requests.

Use sequence analysis to spot suspicious events such as accessing private data and then making big transfers to outside sources.

4:Contextual Correlation:

Review the results from several sources to minimize the number of false positives.

For example, gaining access to the network after work (network log), to customer files (access record) and from the work computer to a personal email would need further examination.

Challenges in Distinguishing Behavior

1:Some people mistakenly identify real actions as suspicious (such as allowing system admins to have access to sensitive information regularly).

2:People’s behavior can change naturally due to projects they are working on, business trips or personal reasons.

3:When systems log data differently or incompletely, it can make it hard to identify the full picture.

4:It is often tricky to recognise where companies’ duties to monitor stop and infringing on privacy starts.

Privacy and Ethical Considerations

1 : How to accord greater security without sacrificing privacy rights.

2 : Pick out only what is strictly necessary and don’t open the documents for inspection when metadata can do the job.

3 : Set the investigation scope and only use the data for that reason.

4: Whenever it is practical, analyze the findings independently of any individuals who contributed them.

5: Oversight: Form an ethics panel that consists of people from both HR and legal departments to supervise the investigation.

6 : Explain to employees the policies of monitoring and outline what is being watched and the reasons behind it.


Maintaining Ethical Standards

1 : Keep records that fully explain all the decisions and rationales used in analysis.

2 : Checks should be added so that analyses are not directed disproportionately at certain groups.

3 : Adjust the investigation to cover the amount and kind of threat involved.

4 : When an employee is flagged, you should make sure to provide a clear explanation.

Communication Strategies

1 : For technical stakeholders (IT security team):

- Present statistics with an explanation of how certain your findings are.

- Provide the findings and core methodology when discussing raw anomaly scores.

- Allow users to view the basic data the model is based on

2 : For non-technical stakeholders (executives, HR):

- Consider the extent of the risk rather than studying technology in detail

- Use charts to display how patterns develop with the passing of time

- Present the outcomes in terms of their relevance to the business.

- Suggest workable options for handling the business issue

3 : For affected employees (if investigation proceeds):
- Discuss your notes without describing the exact ways in which you discovered them

- Give someone an opportunity to explain their behavior.

Question 2 : I've recreated the visitors data visualization shown in the image using Python with matplotlib and seaborn. Here's the detailed implementation:



```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Create the data
data = {
    'Month': ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019'],
    'Searches': [70, 53, 47, 50, 39],
    'Direct': [90, 89, 42, 56, 51],
    'Social Media': [62, 51, None, None, None]  # Last two months missing social media data
}

df = pd.DataFrame(data)

# Set up the plot
plt.figure(figsize=(10, 6))
plt.title('Monthly Website Visitors by Channel', pad=20, fontsize=14)

# Plot each channel
sns.lineplot(data=df, x='Month', y='Searches', marker='o', label='Searches', linewidth=2.5)
sns.lineplot(data=df, x='Month', y='Direct', marker='o', label='Direct', linewidth=2.5)
sns.lineplot(data=df, x='Month', y='Social Media', marker='o', label='Social Media', linewidth=2.5)

# Customize the y-axis
plt.ylim(0, 100)
plt.ylabel('Visitors (in thousands)', labelpad=15)
plt.yticks([0, 20, 40, 60, 80, 100])

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust legend
plt.legend(title='Channels', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
